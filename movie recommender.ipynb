# Remove unwanted columns

```python
# Columns we need:
#     genres
#     id
#     keywords
#     overview
#     title_x
#     cast
#     crew
```

```python
movies = movies[['id','genres','keywords','overview','title_x','cast','crew']]
```

# We have to convert it into a list

```python
import ast
```

```python
def convert(obj):
    ls = []
    for i in ast.literal_eval(obj):
        ls.append(i['name'])
    return ls
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.iloc[0]
```

```python
def convert2(obj):
    ls = []
    count=0
    for i in ast.literal_eval(obj):
        count+=1
        if(count==6):
            break
        ls.append(i['name'])
    return ls
```

```python
movies['cast'] = movies['cast'].apply(convert2)
```

# Have to find director and producer from crew

```python
def convert3(obj):
    ls = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director' or i['job']=='Producer':
            ls.append(i['name']) 
    return ls
```

```python
movies.head(2)
```

```python
movies['crew'] = movies['crew'].apply(convert3)
```

```python
movies.head(2)
```

```python
movies['crew']
```

# Convert overview column into a list

```python
movies['overview'] = movies['overview'].apply(lambda i:i.split())
```

```python
movies.head()
```

 # Now, in all the lists, in each list item, we have to remove the space
 ## Sam Worthington => SamWorthington

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head(2)
```

# Now, we have to concatenate the columns - genres, keywords, overview, cast and crew and create one single column called tags

```python
movies['tags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] +movies['crew']
```

```python
movies['tags']
```

```python
movies.head(2)
```

```python
movies = movies.drop(['keywords','overview','cast','crew','genres'],axis=1)
```

```python
movies.head(2)
```

```python
movies['tags'] = movies['tags'].apply(lambda x:" ".join(x))
```

```python
movies.head(2)
```

```python
movies['tags'][0]
```

```python
movies['tags'] = movies['tags'].apply(lambda x:x.lower())
```

```python
movies['tags'][0]
```

# Will apply stemming to remove similar words

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(txt):
    ls=[]
    for i in txt.split():
        ls.append(ps.stem(i))
    return " ".join(ls)

movies['tags'] = movies['tags'].apply(stem)
```

# Now we will remove stop words

```python
from sklearn.feature_extraction.text import CountVectorizer 

cv = CountVectorizer(max_features=7500,stop_words='english')
```

```python
vectors = cv.fit_transform(movies['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names()
```

# Now we will find cosine similarity for every movie

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:7]
```

# Now we will find 6 most similar movies

```python
def recommender(movie):
    movieIndex = movies[movies['title']==movie].index[0]
    distances = similarity[movieIndex]
    moviesList = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:7]
    
    for i in moviesList:
        print(movies.iloc[i[0]].title)
```

```python
recommender('Avatar')
```
